Women's Health Survey

In 1985, the USDA commissioned a study of women’s nutrition. Nutrient intake was measured for a random sample of 737 women aged 25-50 years. The following variables were measured:

- Calcium(mg)
- Iron(mg)
- Protein(g)
- Vitamin A(μg)
- Vitamin C(mg)

# Descriptive Statistics
We are going to analyze this dataset on the basis of descriptive statistics functions that will help us in understanding the pattern and behavior of the data points.

```python
#Install the library

#!pip install pandas
```

```python
# Import all the Important Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset
data = pd.read_csv("nutrient.csv")
```

```python
data.head(2)
```

```python
print("Columns Names: ")
print(data.columns)

print("\nData Types: ")
print(data.dtypes)
```

```python
data.describe()
```

```python
#Measure of Central Tendency
print("\nMean: ")
print(data.mean())

print("\nMedian: ")
print(data.median())

print("\nMode: ")
print(data.mode())
```

```python
#Measure of Dispersion

print("\nVariance: ")
print(data.var())

print("\nStandard Deviation: ")
print(data.std())

print("\nRange: ")
print(data.max() - data.min())
```

```python
#Skewness and Kurtosis

print("\nSkewness: ")
print(data.skew())

print("\nKurtosis: ")
print(data.kurtosis())
```

```python
#Graph Analysis: Bar Chart

plt.figure(figsize=(6,4))
plt.bar(data['id'], data['calcium'], color = 'blue', label = 'calcium')
plt.xlabel("ID")
plt.ylabel("Calcium")
plt.title("Calcium Intake by ID")
plt.legend()
plt.show()
```

```python
#Graph Analysis: Histogram

plt.figure(figsize=(6,4))
plt.hist(data['calcium'], bins = 10, color = 'red', edgecolor = 'brown')
plt.xlabel("ID")
plt.ylabel("Calcium")
plt.title("Histogram of Calcium Intake by ID")
plt.legend()
plt.show()
```
